```python
data = pd.read_csv("/content/drive/MyDrive/project_I.csv")
data
```

```python
# Create a 2D matrix using User ID (UID) and Product ID
user_product_matrix = data.pivot_table(index='uid', columns='product', fill_value=0)
user_product_matrix
```

```python
# Step 2: Compute features for each user
user_features = user_product_matrix.apply(lambda row: row.astype(bool).sum(), axis=1)
user_features
```

```python
# Step 3: Use a nearest neighbor algorithm
from sklearn.neighbors import NearestNeighbors
```

```python
# Instantiate the Nearest Neighbors model
knn_model = NearestNeighbors(metric='cosine', algorithm='brute')
knn_model.fit(user_product_matrix)
```

```python
# Step 4 and 5: Generate recommendations for each user
recommendations = {}

for user_id in user_product_matrix.index:
    user_vector = user_product_matrix.loc[user_id].values.reshape(1, -1)

    # Find the 6 nearest neighbors (including the user itself)
    _, neighbor_indices = knn_model.kneighbors(user_vector, n_neighbors=6)

    # Extract product IDs from the user and its neighbors
    products = set(user_product_matrix.columns[user_product_matrix.loc[user_id] > 0])
    for neighbor_index in neighbor_indices.flatten():
        if neighbor_index != user_id:
            products.update(user_product_matrix.columns[user_product_matrix.iloc[neighbor_index] > 0])

    # Store the recommendations in the dictionary
    recommendations[user_id] = list(products)
```

```python
# Display the recommendations
for user_id, recommended_products in recommendations.items():
    print(f"User {user_id}: {recommended_products}")
```
